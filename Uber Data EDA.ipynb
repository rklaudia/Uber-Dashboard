```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import time
import pickle
import os

# Load your dataset
# data = pd.read_csv("your_file.csv")

# Get distinct locations
locations = data['Pickup Location'].dropna().unique()

# Setup cache file
cache_file = "location_cache.pkl"
if os.path.exists(cache_file):
    with open(cache_file, "rb") as f:
        cache = pickle.load(f)
else:
    cache = {}

geolocator = Nominatim(user_agent="my_unique_app_name_123")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2, max_retries=5)

results = []
for loc in locations:
    if loc in cache:
        results.append((loc, cache[loc]))
        continue
    try:
        location = geocode(f"{loc}, NCR, India")
        if location:
            cache[loc] = (location.latitude, location.longitude)
            results.append((loc, cache[loc]))
        else:
            cache[loc] = (None, None)
            results.append((loc, (None, None)))
    except Exception as e:
        print(f"Error for {loc}: {e}")
        cache[loc] = (None, None)
        results.append((loc, (None, None)))
    time.sleep(1)  # avoid rate limit

# Save cache
with open(cache_file, "wb") as f:
    pickle.dump(cache, f)

# Convert to DataFrame
coord_df = pd.DataFrame(results, columns=['Pickup Location', 'Coordinates'])
coord_df['Latitude'] = coord_df['Coordinates'].apply(lambda x: x[0])
coord_df['Longitude'] = coord_df['Coordinates'].apply(lambda x: x[1])

# Merge with original data
data_with_coords = data.merge(coord_df[['Pickup Location','Latitude','Longitude']], on='Pickup Location', how='left')

print(data_with_coords.head())
```
